### Prepare the Data

```python
# Identify non-numeric columns
non_numeric_columns = df_market_data.select_dtypes(exclude=['int', 'float']).columns

# Option 1: Convert non-numeric columns (example with one-hot encoding for a categorical column)
df_market_data = pd.get_dummies(df_market_data, columns=non_numeric_columns)

# Option 2: Drop non-numeric columns
# df_market_data = df_market_data.drop(non_numeric_columns, axis=1)


# Assuming df_market_data is your DataFrame
scaled_data = StandardScaler().fit_transform(df_market_data)
```

```python
# Create a DataFrame with the scaled data
df_market_data_scaled = pd.DataFrame(
    scaled_data,
    columns=df_market_data.columns
)

# Copy the crypto names from the original data
df_market_data_scaled["coin_id"] = df_market_data.index

# Set the coinid column as index
df_market_data_scaled = df_market_data_scaled.set_index("coin_id")

# Display sample data
df_market_data_scaled.head()
```

### Find the Best Value for k Using the Original Data.

```python
# Create a list with the number of k-values from 1 to 11
k = range(1,12)
list(k)
```

```python
# Create an empy list to store the inertia values
inertia = []
```

```python
# use a loop to compute the inertia with each possible value of k

# "Inertia measures how well a dataset was clustered by K-Means. It is calculated by measuring the distance between each data point and its centroid, 
# squaring this distance, and summing these squares across one cluster."- Codecademy Inertia is equivalent to the sum of the cluster intra variances.

# Inside the loop, we:
# 1. create a KMeans model using the loop counter for the n_clusters
# 2. fit the model to the scaled data
# 3. append the model.inertia_ to the inertia list


for i in k:
    model = KMeans(n_clusters=i)
    model.fit(df_market_data_scaled)
    inertia.append(model.inertia_)
    
inertia
```

```python
# Create a dictionary with the data to plot the Elbow curve
dic_inertia = {'k':k, 'inertia':inertia}

# Create a DataFrame with the data to plot the Elbow curve
inertia_df = pd.DataFrame(dic_inertia)

inertia_df
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
original_inertia_plot=inertia_df.hvplot(
    x      = 'k',
    y      = 'inertia',
    title  = 'Standarized Performance Data: Inertia versus Number of Clusters',
    xlabel = 'Number of Clusters (k)',
    ylabel = 'Inertia (units)'
)
original_inertia_plot
```

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce to 
# three principal components.
market_data_pca = pca.fit_transform(df_market_scaled)
# View the first five rows of the DataFrame. 
market_data_pca[:5]
```

```python
# Retrieve the explained variance to determine how much information 
# can be attributed to each principal component.
pca.explained_variance_ratio_
```

```python
# Total explained variance with 3 components
pca.explained_variance_ratio_.sum()
```

```python
# Creating a new DataFrame with the PCA data
df_market_data_pca = pd.DataFrame(market_data_pca, columns=['PC1','PC2','PC3'])

# Copy the crypto names from the original data
df_market_data_pca['coin_id'] = df_market_data_scaled.index

# Set the coin_id column as index
df_market_data_pca = df_market_data_pca.set_index('coin_id')

# Display sample data
df_market_data_pca.head()
```

### Find the Best Value for k Using the PCA Data

```python
# Create a list with the number of k-values from 1 to 11
k= list(range(1,12))
k
```

```python
# Initialize an empty list for inertia values
inertia_pca = []

# Loop over a range of k values (1 to 10)
for i in range(1, 11):
    # Create a KMeans model for each k
    model_pca = KMeans(n_clusters=i)
    # Fit the model to the PCA-transformed data
    model_pca.fit(df_market_data_pca)
    # Append the inertia of the model to the inertia list
    inertia_pca.append(model_pca.inertia_)

# Create a dictionary with k values and corresponding inertia
elbow_curve_pca_dict = {'k': range(1, 11), 'inertia': inertia_pca}

# Create a DataFrame from the dictionary
elbow_curve_pca_df = pd.DataFrame(elbow_curve_pca_dict)

# Display the first few rows of the DataFrame
elbow_curve_pca_df.head()
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
elbow_curve_pca_df.hvplot.line(
    title = "Elbow Curve PCA",
    x = "k",
    y = "inertia"
)
```

#### Answer the following question: 

  * **Question:** After visually analyzing the cluster analysis results, what is the impact of using fewer features to cluster the data using K-Means?

  * **Answer:** 
  Principal Component Analysis (PCA) presents a strategic trade-off: it simplifies complex, multi-dimensional data by reducing the number of features, yet still manages to retain a significant portion of the original information. In our case, PCA enabled us to capture 89.5% of the data's variance, effectively distilling multi-dimensional data into a more interpretable two-dimensional format. This enhanced clarity is particularly evident in how we discern correlations among samples.

The PCA plot distinctly illustrates the cluster correlations, making it easier to differentiate between them. Conversely, without PCA, these distinctions become muddled. A prime example is the perceived correlation between clusters 1 and 2 in the non-PCA plot, which, in reality, does not exist. PCA, therefore, not only simplifies data but also clarifies underlying patterns that might be obscured in higher-dimensional spaces.
